```python
import os
import numpy as np
import pandas as pd
```

```python
dirs = '/public/workspace/ryrl/projects/classmates/ryrl/Cancers/TCGA/Counts'
deg = pd.read_csv('/public/workspace/ryrl/projects/classmates/ryrl/Cancers/TCGA/DESeq2_TCGAGTEx.txt', sep='\t', header=0)
deg = deg.assign(Group=deg.apply(
    lambda x: 'Up' if x['log2FoldChange'] > np.log2(2) and x['padj'] < .05 else 'Down' if x['log2FoldChange'] < -np.log2(2) and x['padj'] < .05 else 'Normal', axis=1)).query('Group != "Normal"')
deg.head()
```

```python

genes = deg.query('Cancer in ["LUAD", "LUSC"]')['EnsembleID'].unique().tolist()
len(genes)
```
