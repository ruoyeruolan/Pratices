```python
dirs = '/public/workspace/ryrl/projects/classmates/ryrl/Cancers/TCGA/Counts'
opt = '/public/workspace/ryrl/projects/classmates/ryrl/Cancers/TCGA/Count_'

cancer = 'CHOL'
samples = metaInfo.query('Cancer == @cancer & DataBase == "GTEx"').Sample.to_list()
samples = gtex.columns.intersection(samples)
len(samples)
```

```python
df = pd.read_csv(f'{dirs}/{cancer}.txt', sep='\t', header=0, index_col=0)
df.head()
```

```python
df = pd.merge(df, gtex[samples], left_index=True, right_index=True, how='left')
df.shape
```

```python
i = 0
for cancer in cancers:
    # if not os.path.exists(f'{opt}/{cancer}.txt'):
    # print(cancer, i)
    samples = metaInfo.query('Cancer == @cancer & DataBase == "GTEx"').Sample.to_list()
    samples = gtex.columns.intersection(samples)
    # len(samples)
    # if not os.path.exists(f'{opt}/{cancer}.txt'):
    #     print(cancer)
    #     continue
    try:
        df = pd.read_csv(f'{dirs}/{cancer}.txt', sep='\t', header=0, index_col=0)
        df = pd.merge(df, gtex[samples], left_index=True, right_index=True, how='left').fillna(0).astype(int)
        df.to_csv(f'{opt}/{cancer}.txt', sep='\t', header=True, index=True)
        i += 1
    except Exception as e:
        print(cancer, e)
```

```python
import os
import pandas as pd
dirs = '/public/workspace/ryrl/projects/classmates/ryrl/Cancers/TCGA/results/TCGAGTEx'
files = os.listdir(dirs)
lst = []
for file in files:
    df = pd.read_csv(f'{dirs}/{file}', sep='\t', header=0, index_col=None)
    lst.append(df)
df = pd.concat(lst, axis=0).reset_index(drop=True)
df.head()
```

```python
gtex = pd.read_csv('/public/workspace/ryrl/FK/TCGA/RNA-seq/GTEx_Analysis_v10_RNASeQCv2.4.2_gene_reads.gct.gz', sep='\t', header=0, index_col=0, skiprows=2)
gtex.head()
```

```python
genes = gtex[['Description']].reset_index()
genes = genes.assign(GeneId=genes['Name'].str.split('.').str[0])
genes = genes.groupby('GeneId').agg(lambda x: ','.join(x))
dit = genes.to_dict()['Description']
genes.head()
```

```python
df = df.assign(Symbol=df['EnsembleID'].map(dit))
df.head()
```

```python
df.to_csv('/public/workspace/ryrl/projects/classmates/ryrl/Cancers/TCGA/results/DESeq2_TCGAGTEx.txt', sep='\t', header=True, index=False)
```
